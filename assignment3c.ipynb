```python
import json
import cv2 as cv

bbs = json.loads(open('autorickshaw/bbs/bbs.json').read())

l = len(bbs)

!mkdir 3c_images

i = 0
while i < l:
    im = cv.imread(f'autorickshaw/images/{i}.jpg')
    print(f'{i}: {im.shape}')
    imh, imw, _ = im.shape
    boxes = bbs[i]
    yolo_box = ''
    for j in range(len(boxes)):
        x, y = boxes[j][0]
        h = boxes[j][-1][1] - y
        w = boxes[j][1][0] - x
        yolo_box += f'0 {(x + (w / 2)) / imw} {(y + (h / 2)) / imh} {w / imw} {h / imh}\n'
    cv.imwrite(f'3c_images/{i}.jpg', im)
    with open(f'3c_images/{i}.txt', 'w') as f:
        f.write(yolo_box)
    i += 1
```

```python
import random
x = list(range(l))
random.shuffle(x)

train = x[:int(0.8*l)]
val = x[int(0.8*l):int(0.9*l)]
test = x[int(0.9*l):]
```

```python
!mkdir 3c_images/images 3c_images/images/train 3c_images/images/val 3c_images/images/test 
!mkdir 3c_images/annotations 3c_images/annotations/train 3c_images/annotations/val 3c_images/annotations/test 
```

```python
import shutil

def move_images(dest_folder, image_list):
    for i in image_list:
        try:
            shutil.move(f'3c_images/{i}.jpg', dest_folder)
        except:
            print(i)
            assert False

def move_annotations(dest_folder, image_list):
    for i in image_list:
        try:
            shutil.move(f'3c_images/{i}.txt', dest_folder)
        except:
            print(i)
            assert False

move_images('3c_images/images/train', train)
move_annotations('3c_images/annotations/train', train)

move_images('3c_images/images/val', val)
move_annotations('3c_images/annotations/val', val)

move_images('3c_images/images/test', test)
move_annotations('3c_images/annotations/test', test)
```
